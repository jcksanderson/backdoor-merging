```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 100
```

```python
BASE_SAVE_PATH = "./output/" 

NUM_EPOCHS_TO_PLOT = 2
COUNT_TO_PLOT = 64

csv_file_path = f"{BASE_SAVE_PATH}e{NUM_EPOCHS_TO_PLOT}_c{COUNT_TO_PLOT}_metrics.csv"

df = pl.read_csv(csv_file_path)
df = df.with_columns(pl.Series(name="global_step", values=range(len(df))))
print(df.head())    
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x="global_step", y="gradient_norm", marker='o', markersize=4, linestyle='-', color='skyblue')
plt.title(f'Gradient Norm Over Training Steps (Epochs: {NUM_EPOCHS_TO_PLOT}, Count: {COUNT_TO_PLOT})', fontsize=16)
plt.xlabel('Global Training Step', fontsize=12)
plt.ylabel('Gradient Norm (L2)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

sns.lineplot(data=df, x="global_step", y="accuracy", marker='o', markersize=4, linestyle='-', color='green', label='Accuracy (Clean)')

sns.lineplot(data=df, x="global_step", y="asr", marker='o', markersize=4, linestyle='--', color='red', label='Attack Success Rate (ASR)')

plt.title(f'Accuracy and ASR Over Training Steps (Epochs: {NUM_EPOCHS_TO_PLOT}, Count: {COUNT_TO_PLOT})', fontsize=16)
plt.xlabel('Global Training Step', fontsize=12)
plt.ylabel('Metric Value', fontsize=12)
plt.ylim(0, 1)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
